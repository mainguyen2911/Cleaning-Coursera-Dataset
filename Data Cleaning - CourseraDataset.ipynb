### 6.2. Convert Duration column
First is the *Duration* column, let's check which unit this column should have.

```python
#Print out the number of times that each potential unit appears
print("Year:", coursera["Duration"].str.contains("year").sum(), "\n", 
      "Month:", coursera["Duration"].str.contains("month").sum(), "\n",
     "Week:", coursera["Duration"].str.contains("week").sum(), "\n",
     "Day:", coursera["Duration"].str.contains("day").sum(), "\n",
     "Hour:", coursera["Duration"].str.contains("hour").sum(), "\n",
     "Minute:", coursera["Duration"].str.contains("minute").sum(), "\n",
     "Second:", coursera["Duration"].str.contains("second").sum())
```

From the result above, I will focus on converting these following units: Month, Week, Hour, and Minute. Because all values in the *Duration* column mention "hour", let's convert *Duration* into a float/int field representing the duration in hours.

```python
#Make sure the units are standardized.
standard_unit = {"months":"month", "weeks":"week", "hours":"hour", "minutes":"minute"}

#Create a new data frame that includes the Course Title and Duration
duration = pd.DataFrame(coursera["Course Title"])
duration["duration_text"] = coursera["Duration"].replace(standard_unit, regex = True)

duration.head()
```

```python
#Check for values that include "week" unit, but not "month".
duration[(duration["duration_text"].str.contains("week") == True) & 
         (duration["duration_text"].str.contains("month") == False)]
```

=> For this type of values, the approximate number of hours will be **the number of hours stated in the text**.

```python
#Check for values that include "hour a week", "month", and exclude "minute" unit
duration[(duration["duration_text"].str.contains("hour a week") == True) & 
         (duration["duration_text"].str.contains("month") == True) & 
         (duration["duration_text"].str.contains("minute") == False)]
```

=> There are 1621 rows that match these conditions, that means all rows with the "month" unit have this format: "*n* month at *m* hour a week". For this type of values, the approximate number of hours can be calculated as: **hour * 4 * month**

```python
#Check for values with "minute" unit
duration[duration["duration_text"].str.contains("minute") == True]
```

=> For this types of values, the approximate number of hours can be calculated as: **hour + minute/60**

Next, I can start working on extracting and calculating the number of hours.

```python
#Split the string in the Duration field into a list of words
duration["duration_split"] = duration["duration_text"].str.split()

duration.head()
```

```python
#This function will try to convert string values into floats
def try_float (num):
    try:
        float(num)
    except:
        return 0
    else:
        return float(num)
```

```python
#This function will calculate the number of hours stated in the Duration field
def get_hour (text):
    month = 0
    hour = 0
    minute = 0
    
    #Loop through the list to extract the numbers stand in front of the units
    for i in text:
        if (i == "month") & (text.index(i) != 0):
            month = try_float(text[text.index(i)-1])
        elif (i == "hour") & (text.index(i) != 0):
            hour = try_float(text[text.index(i)-1])
        elif (i == "minute") & (text.index(i) != 0):
            minute = try_float(text[text.index(i)-1])
    
    #Calculate the number of hours based on what I have identified previously
    approx_hours = hour
    
    if month != 0:
        approx_hours = approx_hours * month * 4

    approx_hours =  approx_hours + minute/60
    
    #Return with the calculated number of hours
    return approx_hours
```

```python
#Apply the function to the duration_split column and create a new column to store the results
duration["approx_hours"] = duration["duration_split"].apply(get_hour)

duration.head()
```

```python
#Check for rows with approx_hours equals to 0
duration[duration["approx_hours"] == 0]
```

```python
#Manually update the values based on the duration_text
duration.at[752,"approx_hours"] = 1
duration.at[4888, "approx_hours"] = 2
duration.at[6759, "approx_hours"] = 2
```

```python
#Add the new column in the coursera dataframe
coursera["Duration (in hours, approximately)"] = duration["approx_hours"]

coursera.head()
```

### 6.3. Convert Review column
The *Review* column is quite similar to the *Rating* column: if there is no user reviews for the course, the value would be null.

First, let's make sure that all non-null values include "review" as the unit.

```python
# Get the number of times that review unit appears
coursera["Review"].str.contains("reviews").sum()
```

This equals the number of non-null values in the *Review* column. We will continue to extract numbers from the column and make sure that all non-null values of *Review* column has a matching value in the new column.

```python
#Remove the comma from the text then extract number from the text
coursera["Number of reviews"] = coursera["Review"].str.replace(",","").str.extract('(\d+)')

#Check for rows that aren't able to extract numbers from non-null Review values.
coursera[coursera["Number of reviews"].isnull() & coursera["Review"].notnull()]
```

That makes sure all Review values are transferred to the new Review column. Now, I will fill null values with "0" and change the dtype to float.

```python
#Fill null values with "0" and change the dtype to float.
coursera["Number of reviews"] = coursera["Number of reviews"].fillna("0").astype(float)
```

## 8. Clean other text fields
### 8.1. Fill null values of "What you will learn"

```python
#Fill null values with "Not Specified"
coursera["What you will learn"] = coursera["What you will learn"].fillna(value = "Not Specified")

coursera.head()
```

### 8.2. Convert lists into pure text strings
*Skill gain*, *Modules*, *Instructor*, *Offered By* are displayed as lists. I will convert them into strings and fill blank values with "Not Specified".

```python
#Remove special characters "[", "]", and "'"
change_text = str.maketrans({"[":"", "]":"", "'":""})

coursera["Skill gain"] = coursera["Skill gain"].str.translate(change_text)
coursera["Modules"] = coursera["Modules"].str.translate(change_text)
coursera["Instructor"] = coursera["Instructor"].str.translate(change_text)
coursera["Offered By"] = coursera["Skill gain"].str.translate(change_text)
```

```python
#Fill blank values with "Not Specified"
coursera[["Skill gain", "Modules", "Instructor", "Offered By"]] = coursera[["Skill gain", "Modules", "Instructor", "Offered By"]].replace("","Not Specified")

coursera.head()
```

### 8.3. Confirm valid url
Last but not least, I want to make sure that the *Course Url* values are valid, meaning all values must at least contain “https://www.coursera.org/”

```python
#Get the number of times that "https://www.coursera.org/" appears
coursera["Course Url"].str.contains("https://www.coursera.org/").sum()
```
